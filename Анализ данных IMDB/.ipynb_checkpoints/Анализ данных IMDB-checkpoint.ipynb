# 11. Какого жанра фильмов больше всего?
Варианты ответов:
1. Action
2. Adventure
3. Drama
4. Comedy
5. Thriller

```python
df_genre= pd.DataFrame(data.genres.str.split('|').tolist()).stack().reset_index()
df_genre.columns = ['row_index','row2','genre']
df_genre['profit_genre'] = df_genre.row_index.apply(lambda x : data.iloc[x]['profit'])
display(df_genre.pivot_table(
    values='profit_genre',index='genre',aggfunc='count',margins=True
).sort_values(by='profit_genre',ascending=False).head(10))

answer_ls.append(1)
```

# 18. Какой актер снялся в большем количестве высокобюджетных фильмов? (в фильмах где бюджет выше среднего по данной выборке)

1. Tom Cruise
2. Mark Wahlberg 
3. Matt Damon
4. Angelina Jolie
5. Adam Sandler

```python
df_cast = pd.DataFrame(data.cast.str.split('|').tolist()).stack().reset_index()
df_cast.columns = ['row_index','row2','cast']
df_cast['budget_cast'] = df_cast.row_index.apply(lambda x : int(data.iloc[x]['budget']))
df_cast['year'] = df_cast.row_index.apply(lambda x : data.iloc[x]['release_year'])
pivot_cast=df_cast[df_cast.budget_cast > data.budget.mean()].pivot_table(values='budget_cast',index='cast',aggfunc='count',margins=True)
display(pivot_cast.sort_values(by='budget_cast',ascending=False))

answer_ls.append(1)
```
